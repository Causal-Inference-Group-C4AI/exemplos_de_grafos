## Grafo Balke e Pearl

```python
versao_joaos = """5
5
U1 0
U2 0
X 2
Y 2
Z 2
U1 X
U1 Y
U2 Z
X Y
Z X"""

versao_str =  "U1 -> X, U1 -> Y, U2 -> Z, X -> Y, Z -> X"
U1 = "U1"
U2 = "U2"
X = "X"
Y = "Y"
Z = "Z"
tuple_list = [(U1, X), (U1, Y), (U2, Z), (X, Y), (Z, X)]

G = nx.DiGraph()
G.add_edges_from(tuple_list)

node_colors = []
for node in G.nodes():
    if node == "X":
        node_colors.append("yellow")
    elif node == "Y":
        node_colors.append("red")
    else:
        node_colors.append("lightblue")

nx.draw(
    G,
    with_labels=True,        # Show labels on nodes
    node_color=node_colors,  # Color of nodes
    edge_color="gray",       # Color of edges
    node_size=2000,          # Size of nodes
    font_size=12,            # Font size for labels
    arrowsize=20,            # Arrow size for edges
)
```
