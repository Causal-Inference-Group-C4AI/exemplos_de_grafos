## Grafo 1: **Propagação de uma Epidemia**

- `V`: Vacinação

- `H`: Higiene pessoal

- `UI`: Imunidade natural (não observável)
- `M`: Mídia informativa
- `S`: Sensação de segurança
- `P`: Políticas públicas
- `T`: Testagem

### Arestas (lista de tuplas):
```python
[
    ('A', 'D'), ('D', 'C'), ('C', 'D'), ('H', 'D'),
    ('M', 'V'), ('M', 'S'), ('P', 'V'), ('P', 'H'), ('T', 'D')
]
```

### c-components:
- ``UA``: Ambiente é uma variável não observável que afeta a doença.
- ``UI``: Imunidade natural é uma variável não observável relacionada à infecção.

````python
from utils.funcoes import tuple_generate_example
edges = [
    ('UA', 'D'), ('D', 'C'), ('V', 'I'), ('C', 'D'), ('H', 'D'),
    ('M', 'V'), ('M', 'S'), ('P', 'V'), ('P', 'H'), ('T', 'D')
]
tuple_generate_example(edges)
````
